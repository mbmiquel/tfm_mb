```python

dataframes = []

for i in range(1, 10):
    file_name = f"data/microscore/output_{i}.csv"  
    try:
        df_1 = pd.read_csv(file_name)  
        dataframes.append(df_1)       
    except FileNotFoundError:
        print(f"Fichero {file_name} no encontrado")
    except Exception as e:
        print(f"Error cargando {file_name}: {e}")

if dataframes:
    combined_df = pd.concat(dataframes, ignore_index=True)
    print("Correcto")
else:
    print("Error")
```

```python
combined_df=combined_df[['ID', 'University_Distance', 'School_Distance', 'Kindergarten_Distance',
       'City Center_Distance', 'Supermarket_Distance', 'Bakery_Distance',
       'Hospital_Distance', 'Pharmacy_Distance', 'Restaurant_Distance',
       'Café_Distance', 'Park_Distance', 'Gym_Distance',
       'Movie Theater_Distance', 'Theater_Distance', 'Shopping Mall_Distance',
       'Bus Stop_Distance', 'Metro Station_Distance']]
```

```python
combined_df
```

```python
relevance_weights = {
    "University_Distance": 8,
    "School_Distance": 10,
    "Kindergarten_Distance": 8,
    "City Center_Distance": 10,
    "Supermarket_Distance": 10,
    "Bakery_Distance": 5,
    "Hospital_Distance": 10,
    "Pharmacy_Distance": 8,
    "Restaurant_Distance": 5,
    "Café_Distance": 5,
    "Park_Distance": 5,
    "Gym_Distance": 3,
    "Movie Theater_Distance": 2,
    "Theater_Distance": 2,
    "Shopping Mall_Distance": 7,
    "Bus Stop_Distance": 8,
    "Metro Station_Distance": 10,
}
```

```python
def compute_score(row):
    total_score = 0
    for feature, weight in relevance_weights.items():
        if feature in row.index:
            distance = row[feature]
            if pd.isnull(distance):  
                feature_score = 0
            else:
                # Calcular puntuacion
                feature_score = weight * (0.8 ** max(distance, 0))
            total_score += feature_score
    return total_score

# Apply the scoring function to each row
combined_df['Score'] = combined_df.apply(compute_score, axis=1)
```

```python
def porcentaje_nulos(df):

    # Calcular el porcentaje de nulos
    porcentaje = (df.isnull().sum() / len(df)) * 100

    resultado = pd.DataFrame({
        'Columna': porcentaje.index,
        '% Nulos': porcentaje.values
    }).sort_values(by='% Nulos', ascending=False).reset_index(drop=True)
    
    return resultado
```
